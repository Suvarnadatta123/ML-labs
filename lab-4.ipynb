```python
import pandas as pd

# Step 1: Read CSV file
def read_data(filename):
    try:
        df = pd.read_csv(filename)
        return df
    except FileNotFoundError:
        print("Error: File not found. Please check the file name/path.")
        return None

# Step 2: Find-S Algorithm
def find_s_algorithm(df):
    # Features and target
    X = df.iloc[:, :-1].values
    y = df.iloc[:, -1].values

    # Step 3: Initialize hypothesis with first positive example
    hypothesis = None
    for i in range(len(y)):
        if y[i] == 1 or y[i] == 'Yes':   # handle both numeric and string labels
            hypothesis = list(X[i])
            break

    # If no positive example found
    if hypothesis is None:
        return "No positive example found"

    # Step 4: Generalize hypothesis
    for i in range(len(X)):
        if y[i] == 1 or y[i] == 'Yes':
            for j in range(len(hypothesis)):
                if hypothesis[j] != X[i][j]:
                    hypothesis[j] = '?'

    return hypothesis

# Step 5: Display hypothesis
def display_hypothesis(hypothesis):
    print("Learned Hypothesis:")
    
    if isinstance(hypothesis, list):
        print(" AND ".join(str(attr) for attr in hypothesis))
    else:
        print(hypothesis)

# Step 6: Main function
def main():
    filename = 's-algo.csv'   # your dataset file
    df = read_data(filename)

    if df is not None:
        hypothesis = find_s_algorithm(df)
        display_hypothesis(hypothesis)

# Run program
if __name__ == "__main__":
    main()
```
